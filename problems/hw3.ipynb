Рассмотрим с.в. $\xi\sim \text{Exp}(\lambda)$, имеющую экспоненциальное распределение с параметром $\lambda$:
    $$
        f_{\xi}(x)=\begin{cases}
            \lambda e^{-\lambda x},\quad &x\geq 0\\
            0,&x<0
        \end{cases}
    $$
Пусть $\eta=\operatorname{arctg}\xi$.
    

- Найдите $f_{\eta}(x)$ - плотность распределения случайной величины $\eta$. 
    
- Проверте с помощью численного моделирования на языке **Python**, что искомая плотность была правильной найдена для $\lambda=0.5$. Для этого с помощью функции **numpy.random.exponential** получите $n$ реализаций случайной величины $\eta$ и постройте гистограмму её значений. Сравните полученную гисторамму с теоретической плотностью путём наложения двух графиков.


```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


def exp_dencity(x: np.array, lamb: float):
    assert np.all(x >= 0)
    return lamb * np.exp(-lamb * x)


n = 5000
lamb = 0.5
xi = np.random.exponential(1 / lamb, n)
sns.histplot(xi, stat="density", label="Empirical Exp($0.5$) dencity", bins=30)
x = np.linspace(0, 12, 100)
plt.plot(x, exp_dencity(x, lamb=lamb), label="True Exp($0.5$) dencity", c="tab:red")
plt.legend()
```
